## Exercice 1 : Visualisation de donn√©es simples avec K-Means

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# G√©n√©rer des donn√©es al√©atoires
np.random.seed(42)
X1 = np.random.normal(loc=[1, 1], scale=0.5, size=(100, 2))  # Cluster 1
X2 = np.random.normal(loc=[6, 6], scale=0.5, size=(100, 2))  # Cluster 2
X3 = np.random.normal(loc=[10, 1], scale=0.5, size=(100, 2))  # Cluster 3
X = np.vstack([X1, X2, X3])  # Empiler les donn√©es des 3 clusters

# Appliquer K-means
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_
centers = kmeans.cluster_centers_

# Visualisation des clusters et des centres
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o')
plt.scatter(centers[:, 0], centers[:, 1], marker='X', color='red', s=200, label='Centres des clusters')
plt.title('Clustering K-Means (3 clusters)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()

# Affichage du nombre de points par cluster
unique, counts = np.unique(labels, return_counts=True)
for i, count in zip(unique, counts):
    print(f"Cluster {i} : {count} points")

# Affichage des centres des clusters
print("Centres des clusters :")
print(centers)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# G√©n√©ration d'un ensemble de points en 2D avec des clusters plus proches
np.random.seed(42)
X1 = np.random.normal(loc=[2, 2], scale=0.8, size=(100, 2))  # Cluster 1
X2 = np.random.normal(loc=[4, 4], scale=0.8, size=(100, 2))  # Cluster 2
X3 = np.random.normal(loc=[6, 2], scale=0.8, size=(100, 2))  # Cluster 3
X = np.vstack([X1, X2, X3])  # Empiler les donn√©es des 3 clusters

# Appliquer K-Means avec 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_
centers = kmeans.cluster_centers_

# Visualisation des clusters et des centres
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o')
plt.scatter(centers[:, 0], centers[:, 1], marker='X', color='red', s=200, label='Centres des clusters')
plt.title('Clustering K-Means (3 clusters proches)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()

# Affichage du nombre de points par cluster
unique, counts = np.unique(labels, return_counts=True)
for i, count in zip(unique, counts):
    print(f"Cluster {i} : {count} points")

# Affichage des centres des clusters
print("Centres des clusters :")
print(centers)
```

### **Questions :**

#### 1. **Combien de points appartiennent √† chaque cluster ?**
- **Cas 1 : Clusters espac√©s**  
  Chaque cluster contient environ **100 points**, car les donn√©es sont g√©n√©r√©es de mani√®re relativement uniforme. Les r√©sultats de K-Means devraient √™tre assez √©quilibr√©s, avec des clusters distincts.
  
- **Cas 2 : Clusters plus proches**  
  Les points dans chaque cluster peuvent ne pas √™tre exactement **100**, en raison de la proximit√© des clusters et de l'assignation des points par K-Means. Certains clusters pourraient avoir un nombre l√©g√®rement sup√©rieur ou inf√©rieur de points, mais en g√©n√©ral, la r√©partition est relativement proche de 100 points par cluster.

#### 2. **Que remarquez-vous concernant la position des centres des clusters ?**
- **Cas 1 : Clusters espac√©s**  
  Les centres des clusters sont bien s√©par√©s, chacun √©tant proche du centre de son propre groupe de points. K-Means parvient √† bien s√©parer les groupes en fonction de leurs caract√©ristiques, avec des centres nettement distincts.

- **Cas 2 : Clusters plus proches**  
  Les centres des clusters sont plus rapproch√©s les uns des autres, ce qui peut rendre plus difficile la s√©paration nette entre les clusters. K-Means place les centres l√† o√π la distance entre les points est la plus faible, ce qui peut entra√Æner une certaine "fusion" de clusters si les points sont trop proches les uns des autres.


## Exercice 2 : Variation du nombre de clusters (kk)


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# G√©n√©ration d'un ensemble de points en 2D avec des clusters plus proches
np.random.seed(42)
X1 = np.random.normal(loc=[2, 2], scale=0.8, size=(100, 2))  # Cluster 1
X2 = np.random.normal(loc=[4, 4], scale=0.8, size=(100, 2))  # Cluster 2
X3 = np.random.normal(loc=[6, 2], scale=0.8, size=(100, 2))  # Cluster 3
X = np.vstack([X1, X2, X3])  # Empiler les donn√©es des 3 clusters

# Liste des valeurs de k √† tester
k_values = [2, 3, 4]

# Initialisation de la figure pour les graphiques
plt.figure(figsize=(15, 5))

# Application de K-Means pour chaque valeur de k
for i, k in enumerate(k_values, 1):
    # Application de K-Means
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    labels = kmeans.labels_
    centers = kmeans.cluster_centers_

    # Subplot pour chaque valeur de k
    plt.subplot(1, 3, i)
    plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o')
    plt.scatter(centers[:, 0], centers[:, 1], marker='X', color='red', s=200, label='Centres des clusters')
    plt.title(f'K-Means (k={k})')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.legend()

# Affichage des graphiques
plt.tight_layout()
plt.show()
```

### **Questions :**

#### 1. **Quel k semble le plus adapt√© ? Pourquoi ?**
- **k=3** semble √™tre le plus adapt√©, car il correspond mieux aux trois clusters naturels pr√©sents dans les donn√©es. Avec cette valeur de **k**, K-Means parvient √† bien s√©parer les points en trois groupes distincts, ce qui refl√®te mieux la structure r√©elle des donn√©es.

#### 2. **Quels probl√®mes peuvent survenir si k est mal choisi ?**
- **k trop petit (par exemple, k=2)** : Les clusters naturels peuvent √™tre fusionn√©s en un nombre trop faible de groupes, ce qui entra√Æne une perte d'information et des segments non repr√©sentatifs des donn√©es.
- **k trop grand (par exemple, k=4)** : Le mod√®le peut sur-segmenter les donn√©es, cr√©ant des clusters artificiels qui ne correspondent pas aux groupes r√©els. Cela peut entra√Æner une perte de g√©n√©ralisation et une mauvaise interpr√©tation des donn√©es.


## Exercice 3 : Segmentation des clients avec un petit dataset


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd

# G√©n√©ration d'un petit dataset fictif avec des donn√©es d'√Çge et de D√©penses
np.random.seed(42)
age = np.random.randint(18, 65, 300)  # √Çge des clients entre 18 et 65 ans
depenses = np.random.randint(100, 1000, 300)  # D√©penses entre 100 et 1000

# Cr√©er un DataFrame avec les donn√©es
data = pd.DataFrame({
    '√Çge': age,
    'D√©penses': depenses
})

# Application de K-Means avec k=3
kmeans = KMeans(n_clusters=3, random_state=42)
data['Cluster'] = kmeans.fit_predict(data[['√Çge', 'D√©penses']])

# Visualisation des clusters
plt.figure(figsize=(8, 6))
plt.scatter(data['√Çge'], data['D√©penses'], c=data['Cluster'], cmap='viridis', marker='o')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', marker='X', label='Centres des clusters')
plt.title('Segmentation des clients par √Çge et D√©penses (k=3)')
plt.xlabel('√Çge')
plt.ylabel('D√©penses')
plt.legend()
plt.show()
```

### **Questions :**

#### 1. **Quels groupes peuvent √™tre identifi√©s ?**
- Avec **k=3**, trois groupes distincts de clients sont identifi√©s. Les groupes sont r√©partis selon l'**√Çge** et les **D√©penses** des clients. Le clustering permet de s√©parer les clients en trois segments, bien que la s√©paration exacte d√©pende de la distribution des donn√©es.

#### 2. **Quelles hypoth√®ses pouvez-vous tirer par rapport au comportement des clients de chaque groupe ?**
- **Groupe 1** : Clients jeunes ayant des d√©penses plus faibles, ce qui pourrait refl√©ter un pouvoir d'achat limit√© ou un comportement de consommation plus restreint.
- **Groupe 2** : Clients d'√¢ge moyen, avec des d√©penses mod√©r√©es √† √©lev√©es. Ce groupe pourrait correspondre √† des clients ayant un revenu stable mais pas encore maximal.
- **Groupe 3** : Clients plus √¢g√©s ayant des d√©penses √©lev√©es. Ce groupe pourrait refl√©ter des clients plus √©tablis financi√®rement, avec une capacit√© d'achat plus grande.

### Conclusion :
- K-Means avec **k=3** a permis de segmenter les clients en groupes distincts bas√©s sur **l'√Çge** et les **D√©penses**. Les hypoth√®ses sur le comportement des clients peuvent √™tre faites selon les caract√©ristiques de chaque groupe.
