## Exercice 1 : Création et Manipulation de DataFrames

```python
import pandas as pd

# Création d'un DataFrame à partir d'un dictionnaire
data = {
    'Nom': ['Alice', 'Bob', 'Charlie', 'David'],
    'Âge': [25, 30, 35, 28],
    'Salaire': [50000, 60000, 75000, 55000],
    'Département': ['RH', 'Tech', 'Finance', 'Marketing']
}
df = pd.DataFrame(data)
# 1. Affichage des 3 premières lignes
print("Les 3 premières lignes:")
print(df.head(3))

# 2. Calcul de l'âge moyen
age_moyen = df['Âge'].mean()
print(f"\nL'âge moyen des employés est: {age_moyen:.1f} ans")

# 3. Filtrage des employés avec un salaire > 55000
salaires_eleves = df[df['Salaire'] > 55000]
print("\nEmployés avec un salaire supérieur à 55000:")
print(salaires_eleves)
```

## Exercice 2 : Analyse Statistique Descriptive


```python
# 1. Statistiques descriptives du DataFrame
print("Statistiques descriptives:")
print(df.describe())

# 2. Salaire moyen par département
salaire_par_departement = df.groupby('Département')['Salaire'].mean()
print("\nSalaire moyen par département:")
print(salaire_par_departement)
```

## Exercice 3 : Visualisation des Données


```python
# Création d'un nouveau fichier CSV avec les données de ventes
import pandas as pd
import numpy as np

# Création des données avec le même nombre d'entrées
n_days = len(pd.date_range(start='2024-10-15', end='2024-12-02', freq='D'))
products = ['Monitor', 'Keyboard', 'Tablet', 'Smartphone', 'Laptop']

data_ventes = {
    'Date': pd.date_range(start='2024-10-15', end='2024-12-02', freq='D'),
    'Produit': np.random.choice(products, size=n_days),
    'Prix_Unitaire': np.random.uniform(100, 2000, n_days).round(2),
    'Quantite': np.random.randint(1, 10, n_days),
    'Coût_Production': np.random.uniform(80, 1500, n_days).round(2),
    'Prix_Vente': np.random.uniform(100, 2500, n_days).round(2)
}

# Création du DataFrame
df_ventes2 = pd.DataFrame(data_ventes)

# Sauvegarde dans un fichier CSV avec chemin relatif
csv_path = '../data/donnees_ventes2.csv'
# Créer le dossier data s'il n'existe pas
import os
os.makedirs('../data', exist_ok=True)
df_ventes2.to_csv(csv_path, index=False)


# Importation de données depuis un fichier CSV
#Supposons que vous ayez un fichier 'donnees_ventes.csv'
df_ventes = pd.read_csv('../data/donnees_ventes2.csv')

#Tâches:
# 1. Gérez les valeurs manquantes
df_ventes.dropna(inplace=True)

# 2. Créez une nouvelle colonne calculée
df_ventes['Marge'] = df_ventes['Prix_Vente'] - df_ventes['Coût_Production']

# 3. Triez le DataFrame par marge décroissante
df_ventes_tries = df_ventes.sort_values('Marge', ascending=False)
```
